- selecting rows and cols
    - rows (slicing)
        - loc - gets us the data using named indexes(when the indexes are not in numeric form a,b,c,d..)
        - iloc - gets us the data using numeric indexes (when indexes are numeric value 0,1,2,3..)
    - cols (dictionary style)
        - df[single] - requires name of column
        - df[multiple] - requires a list of columns (multiple is list of named cols)

```python
# single row (record)
weather.iloc[0]
```

```python
weather.iloc[-1] #last record
```

```python
# records 25 - 49
weather.iloc[25: 50]
```

```python
# records 1 - 15 with first 5 col
weather.iloc[1:16, :5] 
```

```python
# records 1 - 15 with first 5 col
weather.iloc[1:16, [0,1,2,3,4]] 
```

```python
# 15 records with last 3 cols
weather.iloc[:15, -3:] # or weather.iloc[:15,[11,12,13]]
```

```python
# 15 records with last 3 cols
weather.iloc[:15, [-3,-2,-1]] # or weather.iloc[:15, [11,12,13]] 
```

- a simple visualization

```python
weather['Precipitation'].plot(figsize=(10,5), kind='kde')  #kde is like histogram
```

```python
weather['MaxTemp'].plot(figsize=(10,5), kind='kde')
```

```python

weather['MaxTemp'].plot(figsize=(20,6), kind='hist', color='green', bins=25, alpha=.5)  
```

```python
weather[['MaxTemp','MinTemp']].plot(figsize=(20,6), 
    kind='hist', color=['green','hotpink'], bins=25, alpha=.5)  #bins is no. of pieces that can be seen from above
    #alpha is used to adjust transparency of different graphs
```

```python
weather[['MaxTemp','MinTemp']].plot(figsize=(20,8), 
    kind='hist', color=['green','hotpink'], bins=25, alpha=.5, subplots=True, layout=(2,1))
```

```python
weather[['MaxTemp','MinTemp']].plot(figsize=(20,5), 
    kind='hist', color=['green','hotpink'], bins=25, alpha=.5, subplots=True, layout=(1,2))
```
